```python
import os
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt

classes = ['Raiva', 'Alegria', 'Neutro', 'Triste', 'Surpresa']
```

```python
# Função para realizar a previsão e imprimir o resultado
def predict_image(file_path):
    img = image.load_img(file_path, target_size=(200, 200))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = img / 255.0

    prediction = model.predict(img)

    rating_mais_provavel = np.max(prediction)
    emocao = classes[np.argmax(prediction)]
    print("Emoção provável ", emocao, f', %.2f' % (rating_mais_provavel * 100))
    print(prediction)
```

```python

new_image_path = './test/Surpreso2.png'

img=plt.imread(new_image_path)
plt.axis('off')
plt.imshow(img)
predict_image(new_image_path)
```

```python
# Pasta contendo imagens teste
test_folder = './test'
imagens_teste = os.listdir(test_folder)
for image_file in imagens_teste:
    print(f"\nPrevisão para: {image_file}")
    predict_image(test_folder+'/'+image_file)    img_path = test_folder+'/'+image_file
    predict_image(img_path)
```
